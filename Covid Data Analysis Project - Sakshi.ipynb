```python
vaccine_df = pd.read_csv("D:/Exploratory Data Analysis on Covid - 19/covid_vaccine_statewise.csv")
```

```python
vaccine_df.head(7)
```

```python
covid_df.drop(["Sno","Time", "ConfirmedIndianNational","ConfirmedForeignNational"], inplace = True, axis=1)
```

```python
covid_df.head()
```

```python
covid_df["Date"] = pd.to_datetime(covid_df["Date"], format = "%d-%m-%Y")
#corrected error at b399
```

```python
#active cases

covid_df["Active_Cases"] = covid_df["Confirmed"] - (covid_df["Cured"] + covid_df["Deaths"])
covid_df.tail()
```

```python
#creating pivot table using pandas

statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"], 
                          index = "State/UnionTerritory", aggfunc = max)
```

```python
# finding recovery rate
statewise["Recovery_Rate"] = statewise["Cured"]*100/statewise["Confirmed"]

# finding mortality rate
statewise["Mortality_Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
#sorting based on confirmed cases

statewise = statewise.sort_values( by = "Confirmed", ascending = False)
```

```python
#plotting table
statewise.style.background_gradient(cmap = "twilight")
```

```python
#Top 10 state with active case

top_10_cases = covid_df.groupby(by = "State/UnionTerritory").max()[["Active_Cases", "Date"]].sort_values(by = ["Active_Cases"], ascending = False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 States with most Active Case",size = 18)
```

```python
top_10_cases = covid_df.groupby(by = "State/UnionTerritory").max()[["Active_Cases", "Date"]].sort_values(by = ["Active_Cases"], ascending = False).reset_index()
fig = plt.figure(figsize=(16,9))
ax= sns.barplot(data=top_10_cases.iloc[:10], y="Active_Cases", x="State/UnionTerritory")
plt.title("Top 10 States with most Active Case",size = 22)
```

```python
top_10_deaths = covid_df.groupby(by = "State/UnionTerritory").max()[["Deaths", "Date"]].sort_values(by = ["Deaths"], ascending = False).reset_index()
```

```python
#Top 10 state with death
top_10_deaths = covid_df.groupby(by = "State/UnionTerritory").max()[["Deaths", "Date"]].sort_values(by = ["Deaths"], ascending = False).reset_index()
fig = plt.figure(figsize=(18,5))
plt.title("Top 10 States with most Deaths",size = 22)
axe = sns.barplot(data=top_10_deaths.iloc[:12], y="Deaths", x="State/UnionTerritory", linewidth = 2, edgecolor="black")
```

```python
#growth trend

fig= plt.figure(figsize = (12,6))

ax = sns.lineplot(data = covid_df[covid_df["State/UnionTerritory"].isin(["Maharashtra","Karnataka","Kerala","Tamil Nadu","Uttar Pradesh"])], x= "Date", y= "Active_Cases", hue="State/UnionTerritory")
plt.legend(loc='upper left')
myFmt = DateFormatter("%b-%Y")
ax.xaxis.set_major_formatter(myFmt)

## Rotate date labels automatically
fig.autofmt_xdate()
ax.set_title("Top 5 affected states", size= 16)
```

```python
#remove india as state

vaccine = vaccine_df[vaccine_df.State!="India"]
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated": "Total"},inplace = True)
vaccine.head()
```

```python
#state with most least vaccinations

max_vac = vaccine.groupby("State")["Total"].sum().to_frame("Total")
max_vac = max_vac.sort_values("Total",ascending=False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(15,10))
plt.title("TOP 5 Vaccinated States", size = 20)
x= sns.barplot(data = max_vac, y = max_vac.Total, 
               x = max_vac.index, linewidth = 2,
               palette = "gist_earth")
plt.ylabel("Vaccinations")
plt.show()
```

```python
min_vac = vaccine.groupby("State")["Total"].sum().to_frame("Total")
min_vac = min_vac.sort_values("Total",ascending=True)[:5]

fig = plt.figure(figsize=(15,10))
plt.title("Least 5 Vaccinated States", size = 20)
x= sns.barplot(data = min_vac, y = min_vac.Total, 
               x = min_vac.index, linewidth = 2,
               palette = "gist_earth_r")
plt.ylabel("Vaccinations")
plt.show()
```
